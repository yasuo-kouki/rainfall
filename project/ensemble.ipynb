```python
import joblib
import numpy as np

# 保存したモデルを読み込み test学習まで終わっている
model_gbdt = joblib.load('gbdt_model/xgb_model.pkl')
model_deep = np.load("deep_model/test_probs.npy")
```

```python
gbdt_label = np.where(model_gbdt > 0.5 ,1, 0)
deep_label = np.where(model_deep > 0.5, 1 ,0)
```

```python
deep_label
```

```python
gbdt_label
```

```python
# 各モデルの予測確率を加重平均
weight_xgb = 0.5
weight_lr = 0.5
final_pred_probs = model_gbdt * weight_xgb + model_deep * weight_lr

# 最終的なクラス予測（確率が0.5を超えたらクラス1、それ以下はクラス0）
final_pred_label = np.where(final_pred_probs > 0.5, 1, 0)

# 結果を表示
print(f"Final predicted classes: {final_pred_label}")
```

```python
import pandas as pd
test = pd.read_csv('data/test.csv')
# 提出用ファイルの作成
submission = pd.DataFrame({
    'id': test['id'], 
    'rainfall': final_pred_label
})
submission.to_csv('submit/ensembel/submission_2.csv', index=False)
```
